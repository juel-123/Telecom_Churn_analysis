**Analyze the Churn feature**

```python
Churn = pd.DataFrame(df['Churn'].value_counts())
Churn
```

```python
df['Churn'].value_counts().plot.pie(explode = [0.05,0.05],autopct= '%1.1f%%', startangle= 90, shadow = True, figsize=(5,5))
plt.title('Pie chart for Churn')
plt.show()
```

**Analyzing the State column**

```python
#comparing Churn with the State column using countplot
sns.set(style="darkgrid")
plt.figure(figsize=(15,8))
ax = sns.countplot(x = 'State', hue = 'Churn' , data = df )
plt.show()
```

```python
# show the top ten most Churned state
grouped_data = pd.DataFrame( df.groupby("State")["Churn"].mean().sort_values(ascending= False).head(10))
grouped_data
```


**There is 51 unique state who have different churn rate
  , From the above analysis we get to know that  NJ, CA, TX, MD, SC has the highest churn rate,
*The reason for the Churn Rate may be due to the low coverage of cellular network***

**Analyzing "Area code" column**

```python
#calculate Area code vs Churn percentage
Churn_area = pd.DataFrame(df.groupby("Area code")['Churn'].value_counts())
Churn_area['Churn_rate'] = round(Churn_area*100/len(df),2)
Churn_area
```

```python
sns.set(style = "darkgrid")
ax = sns.countplot(x = 'Area code' , hue = "Churn" ,data = df)
plt.show
```

***In the above we notice that there are only 3 unique value of area code i.e: 408,415,510 and the churnrate of these Area code is almost same. we don't think there is any kind of relationship between "Area code" and "Churn" due to which customer leaves the operator.***

**Analyzing "Account length" Column**

```python
# separating Churn and Non Churn customers
Churn_df = df[df['Churn']==bool(True)]
not_Churn_df = df[df['Churn']== bool(False)]
```

```python
#Account length vs Churn
sns.distplot(df['Account length'])
plt.show()
```

```python
#comparsion of churned Account length and Non Churned Account length
sns.distplot(df['Account length'], color= 'yellow', label = 'All')
sns.distplot(Churn_df["Account length"], color= 'red',hist = False ,label = 'churned')
sns.distplot(not_Churn_df["Account length"], color= 'green',hist = False, label = 'not_churned')
plt.legend()
```

***After analyzing various aspects of "Account length" column we didnot find any useful relation to churn, so we are not able to find any connection with the churn as of now . let's see other features***

**Analyzing "International plan" Column**

```python
#show the value_counts of International plan
df['International plan'].value_counts()
```

```python
#calculate International plan vs Chun rate
International_plan_data = pd.DataFrame(df.groupby('International plan')['Churn'].value_counts())
International_plan_data['Churn_Percentage']= round(International_plan_data*100/len(df),2)
International_plan_data
```

```python
# Analyzing using countplot
sns.countplot(x= 'International plan', hue = "Churn", data = df)
plt.show()
```
